#### **（4）复习可变参数与不可变参数内容**
在python中，数据类型的可变与不可变之分的关键在于，当变量修改其值时，其内存地址是否改变，即是否创建新的对象

- 不可变类型（Immutable）
  - **特点：对象创建后状态不可修改**
  - **内存机制**：任何修改操作都会创建新对象
- 可变类型（Mutable）
  - **特点**：对象内容可修改且保持同一 ID
  - **内存机制**：操作直接修改原对象内存区域

数据类型分类与特性对比

| 类型  | 可变性 | 示例  | 内存变化特征 |
| --- | --- | --- | --- |
| 数字(int/float) | 不可变 | 10, 3.14 | 修改会创建新对象 |
| 字符串(str) | 不可变 | “Hello” | 任何修改生成新字符串 |
| 元组(tuple) | 不可变 | (1, “a”, [2]) | 元素引用不可变（但可含可变元素） |
| 列表(list) | 可变  | [1, 2, 3] | 原位修改不改变对象 ID |
| 字典(dict) | 可变  | {“key”: “value”} | 键值对增删改不影响对象地址 |
| 集合(set) | 可变  | {1, 2, 3} | 元素增删不改变集合本体地址 |

```python
#（1）定义一个闭包，用于求解方程的y与x值的变化，例如 y = ax + b。
def linear_equation(a, b):
    def calculate_y(x):
        y = a * x + b
        return y
    return calculate_y

# 定义函数
func = linear_equation(3,4)
x_lst = range(1, 100, 9)
for x in x_lst:
    y = func(x)
    print(f'when x = {x}, then y = {y}')
```

```python
#（2）创建一个闭包，实现统计函数执行的次数功能
def count_func_exec():
    count = 0
    def func():
        nonlocal count
        count += 1
        print('hello world')
        print(f'函数执行了{count}次')
    return func

f = count_func_exec()
f()
f()
f()
```

```python
#（3）请使用装饰器方式来统计输出100000句"程序员YYDS"的执行时间。
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        print(f'函数 {func.__name__} 的执行时间为：{end - start} s')
    return wrapper

@timer
def test():
    for i in range(100000):
        print("程序员YYDS")

test()
```

```python
#（4）定义一个函数, 返回字符串, 使用装饰器实现对这个字符串添加后缀.txt。
import time, uuid

def add_file_suffix(suffix_name: str):
    if suffix_name.startswith(r'.'):
        def decorator(func):
            def wrapper(*args, **kwargs):
                return_str =  func(*args, **kwargs) + suffix_name
                return return_str
            return wrapper
        return decorator
    else:
        raise ValueError('文件后缀必须以"."开头')

@add_file_suffix('.txt')
def generate_tmp_file_name():
    return f'{str(uuid.uuid4()).replace('-', '')}_{time.time_ns()}'

for i in range(100):
    print(generate_tmp_file_name())
```

```python
# (5) 编写一个闭包函数 `outer`，它接受一个参数 `x`，并返回一个内部函数 `inner`。
# `inner` 函数接受一个参数 `y`，并返回 `x + y` 的结果。测试这个闭包。
def outer(x):
    def inner(y):
        return x + y
    return inner

print(outer(1)(2))
print(outer(10)(5))
```

```python
# (6) 编写一个闭包函数 `counter`，它返回一个内部函数 `increment`。 每次调用 `increment` 时，它会增加并返回一个计数器。使用 `nonlocal` 关键字来实现。
def counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment
counter = counter()
print(counter())
print(counter())
print(counter())
print(counter())
```

```python
# (7) 编写一个装饰器 `@uppercase`，它将函数的返回值转换为大写。 参考函数：upper()_转换为大写
def uppercase(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

import uuid
@uppercase
def get_str():
    return f'{str(uuid.uuid4()).lower()}'

print(get_str())
print(get_str())
print(get_str())
print(get_str())
print(get_str())
```

```python
# (8) 编写一个装饰器 `@repeat(n)`，它可以让被装饰的函数重复执行 `n` 次。
def repeat(count:int):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(count):
                yield func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(10)
def count_range(start, end):
    count = 0
    for i in range(start, end):
        count += 1
    return count

for i in count_range(1, 100):
    print(i)
```

```python
# (9) 编写两个装饰器 `@uppercase` 和 `@exclaim`，分别将函数的返回值转换为大写和添加感叹号。测试它们的组合效果。
def uppercase(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

def exclaim(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs) + '!test'
    return wrapper

import uuid
@uppercase
@exclaim
def get_str1():
    return f'{str(uuid.uuid4()).lower()}'

@exclaim
@uppercase
def get_str2():
    return f'{str(uuid.uuid4()).lower()}'

for i in range(10):
    print(f'get_str1 执行结果：{get_str1()}')
    print(f'get_str2 执行结果：{get_str2()}')
```
